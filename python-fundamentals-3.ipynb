```python
# 1. IF STATEMENT
# take user input age
age = int (input("Enter your age:"))
# if the age is greater than 18, print access granted

# if statement starts
if age >=18: # if header to check the condition
    print("Access granted") # indentation is important here! 
    print("Welcome to FPS Game, enjoy!")

print("Program ended")
```

```python
# 2. ELIF STATEMENT
# use if you have two different scenarios to run
# take user input age
age = int (input("Enter your age:"))
# if the age is greater than 18, print access granted

# if statement starts
if age >=18:
    print("Access granted") # indentation is important here! 
elif age <=0:
    print ("Input incorrect") 
```

```python
# 3. ELSE STATEMENT
# Use this if your have more than two scenarios
# take user input age
age = int (input("Enter your age:"))
# if the age is greater than 18, print access granted

# if statement starts
if age >=18:
    print("Access granted") # indentation is important here! 
elif age <=0:
    print ("Input incorrect") 
else:
    print ("Sorry you are too young for this game.")
```

**Python While Loop**
* 


[image.png](attachment:fb9d466f-310d-4f39-80e0-2e9068dc7c3f.png)
1. A while loop evaluates the condition
2. If the condition evaluates to True, the code inside the while loop is executed
3. Condition is evaluated again
4. This process continues until the condition is False
5. When condition evaluates to False, the loop stops.

```python
# WHILE : run the code as long as the condition is true

# assign the secret number the user needs to guess
secret_number = 3

# take the user input
guess = int(input("Guess a number: "))

# condition of the while loop
while guess != secret_number:
    guess = int(input("Guess a number: "))
else:
    print("Congratulations, you guess my number!")
    
# code will run again when the guess is false, it will stop when the guess is correct
```

**While loop with Else**
* Python while loop can have optional else block. The else part is executed after the condition of the loop evaluates to False.

```python
counter = 0

while counter < 10:
    print('Inside loop')
    counter = counter + 1
else:
    print('Inside else')
```

Use 'break' key word to terminate a loop

```python
counter = 0

while counter < 10:
    # loop ends because of break
    # the else part is not executed 
    if counter == 5:
        break

    print('Inside loop')
    counter = counter + 1
else:
    print('Inside else')
```

**For Loop in Python**
* When you want to repeat a statement at the exact value, for example to repeat "Hello" 10 times.

* Use for ... in range()

```python
# range function will generate the values, this always start with 0
# i is the control variable
# the block will run as long as it is in range
for i in range(10):
    print(i)
    
# print hello 5 times
for i in range(5):
    print("Hello")
```

```python
# the range function can accept the starting value and end value-1 arguments
for i in range(1,9):
    print(i)
```

```python
# your turn, print number 25 to 30 using for loop
for i in range(25,31):
    print(i)
```

```python
# Sometime we want to stop the loop when we reach certain number

for i in range(5):
    if (i == 2): # the loop will stop at i = 2, use if statement here
        break # use break key word to do this
    print ("i is: ", i)
```

```python
# Sometime we want to skip the loop when we reach certain number

for i in range(5):
    if (i == 2): # the loop will stop at i = 2, use if statement here
        continue # use continue key word to do this
    print ("i is: ", i)
```

**Summary**
1. if/else statements allow us to conditionally run code

3. Use for...in loop to execute code for each item in a sequence
