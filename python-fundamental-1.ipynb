# Section 2:**Python Operator**
There are seven mathematics operators in python:


```python
# Exponentiation, use double asteriks **
print (2 ** 3) # both integers, result is integer
print (2 ** 3.) # if one of the argument is float, result is float
print (2. ** 3) # result it float
print (2. ** 3.)
print (type(2. ** 3)) # output the type of the exponentiation result
```

Exponentiantion summary:
- when both ** arguments are integers, the result is an integer
- when at least one ** argument is a float, the result is a float

```python
# Multiplication *
print (2 * 6)
print (2 * 6.)
print (2. * 6)
print (2. * 6.)
```

Multiplication summary:
- Multiplication works almost the same with the exponentiation
- When at least one argument is a float number, the result is also a float

```python
# Division /
print(6 / 3)
print(6 / 3.)
print(6. / 3)
print(6. / 3.)
```

Division summary: the result will **always** in floating point number

If we would like to have a whole number result, use floor division (//)

```python
print (6 / 3) # division give float result
print ( 6 // 3) # floor div. give whole number result
```

```python
# floor division round to the lower number
print (5 / 2) # 2.5
print ( 5 // 2) # 2

print ( -5 / 2) # -2.5
print (-5 // 2) # -3

print (-12 / 4) # -3.0
print (-12 // 4) # - 3
```

**Important:** Floor division will round to the lower number!

**Modulo (%)**:
Modulo will result in the remainder of a division operation of two arguments.



```python
#  Modulo
print (7 / 2) # 3.5
print ( 7 % 2 ) # 1
print ( 8 % 2) # 0
print (12 % 4.5) # 12 / 4.5 = 9.0 R 3.0
print (12.1 % 2) # will not round the result
```

**Points to remember about division:**
1. Do not perform division by zero eg: 5/0 , 9.8/0
2. Do not perform an integer division by zero eg : 10/0
3. Do not try to find a remainder of a division by zero

```python
print ( 5/ 0)
```

**Operators and their bindings**
Binding : will determine the order of computations

Most of the python operators have left-side binding

```python
print (9 % 6 % 2) # 3 % 2 = 1, modulo has left-side binding
```

```python
print ( 2 ** 2 ** 3)
```

###### with exponentiation, there are two possible answers:
1. left-side binding : 2 ** 2 = 4 -> 4 ** 3 = 64
2. right-side binding: 2 ** 3 = 8 -> 2 ** 8 = 256

because the result is 256, it means that exponentiation operation uses **right-side binding**

```python
print ( 3 ** 2 ** 2) # 2 ** 2 = 4 -> 3 ** 4 = 81
```

```python
print(-3 ** 2) # -9
print(-2 ** 3) # 8
print(-(3 ** 2)) # -9
```

**List of priorities**




```python
print (2 * 3 % 5) # do the multiplication then modulus
```

```python
print ((5 * (( 25 % 13) + 100) / (2 * 13)) // 2)
```
