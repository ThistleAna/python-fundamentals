# Python List
List is a collection of multiple elements.
For example:

```python
# creating list
countries = ["India", "USA", "United Kingdom"] 
# each element has index, starts at index 0
# printing list
print(countries[0])
#print all list
print ("Original list:", countries)
#changing element on a specific index
countries[0] = "Japan"
print("Modified list:", countries)
```

```python
# use len() to count the number of element inside list
len(countries)
```

```python
# use del to delete element of a specific indeX
age = [23, 45, 67, 78, 23]
print ("Original list : ", age)
print ("Length of the element:", len(age))
del age[3]
print ("New list: ", age)
print ("Length of the element:", len(age))
```

**Negative Indices Are Legal (!)**


```python
# getting the last element of the list use [-1]
name = ["Abdi", "Bob", "Christopher"]
print(name[-1])
# last accesssible element in this list is -3
```

**We cannot access a value that the index don't exist**

# Python Methods
Methods are specific kind of functions. With methods, it only works for the data that we need it to work for unlike function that is a stand alone.



**Manipulating List using Methods**

```python
# list.append() is to add new element to a list
name = ["Abdi", "Bob", "Christopher"]
print(name)
name.append("Debbie")
print(name)
```

```python
# use list.insert(index, element) to add new element at specific index
name = ["Abdi", "Bob", "Christopher"]
print(name)
name.insert(2, "Debbie")
print(name)
```

```python
# Technique to swap value in the list
name = ["Abdi", "Bob", "Christopher"]
print(name)
# swap Bob and Christoper
temp = name[1]
name[1] = name[2]
name[2] = temp
print(name)
```

**We can swap value easily using this trick:**

```python
name = ["Abdi", "Bob", "Christopher"]
name[1], name[2] = name[2], name[1] # the value of the index 1 to swap with index 2 and the value of index 2 to swap with index 1
print(name)
```

Exercise:
Given a list of ages on a particular Adult Education Students: 56, 72, 24 and 46. Iterate through the list and print the average age of the students class.

```python
# create list
ages = [56, 72, 24, 46]
# total variable to store total age
total = 0
# loop through age using for 
for age in ages:
    total += age # is the same as total= total + age
# calculate the average, store inside average variable
average = total/ len(ages)
# print total age
print("Age list:", age)
print("Total age:", total)
print("Average age:", average)
```

**Slicing a List**
To slice a list, we can use two reference 
* 

```python
# slicing a list
letters = ["A", "B", "C", "D", "E"]
# slicing the list, get the first two elements
firstTwo = letters[0:2]
# print the new list
print(firstTwo)
# print all the way to the end
print(letters[1:]) # only specify the starting index
# specify the end index only
print(letters[:3])
```

```python
# Slicing using negative indices
countries = ["Argentina", "Bahrain", "China", "Dominica", " Eritrea"]
# slicing using negative indices
print(countries[1:-1]) # starts from index 1, end one element before the last element on the list
```

```python
# what if we do not specify the indices?
countries = ["Argentina", "Bahrain", "China", "Dominica", " Eritrea"]
# slicing using no indices specified
print(countries[:]) # copy the entire list
```

**Note: when we slicing a list, we are creating new list on the memory**

```python
# using the del keyword
countries = ["Argentina", "Bahrain", "China", "Dominica", " Eritrea"]
# use del 
del countries[1:3] # delete index 1 and 2
print(countries) 
```

**When using del keyword, we are modifying the original array. For example, original length is 5, after using the del the length decrease to 3**

```python
# using the del keyword
countries = ["Argentina", "Bahrain", "China", "Dominica", " Eritrea"]
# use del 
del countries[:] # delete the whole array
print(countries) 
```
