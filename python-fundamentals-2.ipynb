Variable naming:
- must start with a letter or the underscore character
- cannot start with number
- can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
- case sensitive, (name, Name, and NAME are three different variables)
- other than reserved keywords


```python
# valid variable name
score = 10
print(score)
Score = 50
print(Score)
Sc_o_re = 22
print(Sc_o_re)
_score = 34
print(_score)
score_ = 77
print(score_)
_SCORE_ = 100
print("_SCORE_", _SCORE_)
```

**Multi Words Variable Names**
1. Camel Case: each word, except the first, starts with capital letter, myNameIs = "Bob"
2. Pascal Case: each word starts with a capital letter: MyNameIs = "Bob"
3. Snake Case : each word is separated by an underscore character, my_name_is = "Bob"

```python
# Camel case
myNameIs = "Bob"
# Pascal case
MyNameIs = "Charlie"
# Snake case
my_name_is = "Amina"

print (myNameIs, MyNameIs, my_name_is)

studentPerClass = 20 # camel case
amount_per_liter = "snake" # snake case
TheMostBeautifulPlaceInTheWorld=2 # pascal case

print(studentPerClass, amount_per_liter, TheMostBeautifulPlaceInTheWorld)
```

**Variable Assignment in Python**
Use an equal sign to assign value to a variable.
We can update the value by re-declaring the variable with the new value.

```python
age = 12
age = 78
age = 34
print(age)
```

**Python Assign Values to Multiple Variables**
Python allows assigning a single value to several variables simultaneously with "=" operators.



```python
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

**Python Assign Different Values to Multiple Variables**
Use comma "," to assign different values in single line

```python
x, y, z = "Orange", "Banana", "Cherry"
print("x =", x)
print("y =", y)
print("z =", z)
```

**The plus sign**
- It can connect one variable to variable if the variables are string value
- It will add two variables if the values are numbers
- It is recommended to connect string variables with comma instead of plus sign to avoid logical error

```python
first_name = "Harry"
last_name = "Potter"
print (first_name, last_name)

number1 = 30
number2 = 40
print (number1 + number2)

var1 = 40
var2 = "Ball"
print(var1,var2) # if it is separated by + it will give error
```

**Shortcut Operators**

* x = x * 2 is the same as x *= 2
* x = x + 2 is the same as x += 2


```python
x = 50
y = x
y += 5
print ("y =", y)
print ( "x =", x)

x += 2  # shortcut 
print (x)

x -= 2 # shortcut
print (x)
```
